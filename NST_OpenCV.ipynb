# **Sending The INPUT IMAGE**

```python
image = cv2.imread("/content/IMG_20180514_065111.jpg")
image = imutils.resize(image, width=600)
(h, w) = image.shape[:2]
```

```python
image.shape
```

```python
blob = cv2.dnn.blobFromImage(image, 1.0, (w, h),
	(103.939, 116.779, 123.680), swapRB=False, crop=False)
net8.setInput(blob)
output = net8.forward()
```
